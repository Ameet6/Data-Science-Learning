```python
import matplotlib.pyplot as plt
```

```python
x = [80,110,120,150,200]
y = [30,50,60,90,100]
plt.plot(x,y)
plt.title("Price per square yard")
plt.xlabel("Square yard")
plt.ylabel("Price in lacs")
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x,y1,label = 'y = x**2', color = "red")
plt.plot(x,y2,label = 'y = x**3', color = "blue")
plt.title("Multi lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.savefig("Multiline graph")
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]

plt.plot(x,y1,label = 'y = x', color = "red")
plt.plot(x,y2,label = 'y = y', color = "blue")
plt.title("Multi lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.ylim(-200,200)
plt.xlim(-5,5)
plt.axhline(0)
plt.axvline(0)
plt.legend()
plt.savefig("Multiline graph")
plt.show()
```

```python
x = [80,110,120,150,200]
y = [30,50,60,90,100]
plt.plot(x,y,color = 'green', linestyle = 'dashed',linewidth = 1, marker ='o',markerfacecolor = 'red',markersize = 6)
plt.title("Price per square yard")
plt.xlabel("Square yard")
plt.ylabel("Price in lacs")
plt.grid()
plt.show()
```

```python
Grades = ['A','B','C','D']
Students = [30,50,20,10]
plt.bar(Grades,Students,edgecolor = 'red',color = ['red','blue','green','orange'])
plt.xlabel("Student Grades")
plt.ylabel("No of Students")
plt.title("Grade Distribution")
plt.show()
```

```python
Grades = ['A', 'B', 'C', 'D']
Students = [30, 50, 20, 10]

plt.bar(Grades, Students,facecolor = 'lightgreen', label="Student Grades")
plt.legend()  # Shows 'Student Grades' in a box
plt.title("Grade Distribution")
plt.show()
```
