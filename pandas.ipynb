# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\HP\Downloads\Customer Call List.xlsx')
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# df['Last_Name'] = df['Last_Name'].str.lstrip("....")
# df['Last_Name'] = df['Last_Name'].str.lstrip("///")
# df['Last_Name'] = df['Last_Name'].str.rstrip("_ _")
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df
```

```python
# df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:10]}" if len(x) == 10 else x)
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:10]}" if len(x) == 10 else x)
df
```

```python
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',n = 2,expand = True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df = df.replace('N/a','')
# df = df.replace('NaN','')
df = df.fillna('')
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
# # df = df.dropna(subset = "Phone_Number", inplace = True)
# df = df[df['Phone_Number'] != '']
```

```python
df
```

```python
df.reset_index()
```

```python
df.reset_index(drop = True)
```
